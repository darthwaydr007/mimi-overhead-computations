```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
font = {
    'size'   : 16
}
rc('font', **font)
# 1. Load the data
df = pd.read_csv("data.csv")

# 2. Convert times from nanoseconds to milliseconds and bytes to KB
df["AppSendTime_ms"]   = df["AppSendTime_ns"]   / 1e6
df["ProofCompTime_ms"] = df["ProofCompTime_ns"] / 1e6
df["LocalMem_KB"]      = (df["LocalMem"] / 1024).round(2)

# 3. Plot application send vs. proof computation times
plt.figure(figsize=(8, 4))
plt.plot(df["GroupSize"], df["AppSendTime_ms"],   marker="o", label="App send time")
plt.plot(df["GroupSize"], df["ProofCompTime_ms"], marker="s", label="Proof computation")
plt.xlabel("Group size (number of users)")
plt.ylabel("Time (milliseconds)")
#plt.title("Application Send Time & Merkle-Proof Computation Time\nacross Group Sizes")
plt.legend(loc="upper left")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()

# 4. Plot additional local memory overhead
plt.figure(figsize=(8, 4))
plt.plot(df["GroupSize"], df["LocalMem_KB"], marker="^")
plt.xlabel("Group size (number of users)")
plt.ylabel("Extra local memory (KB)")
##plt.title("Local Memory Overhead for Merkle-Proof Processing\nby Group Size")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```
